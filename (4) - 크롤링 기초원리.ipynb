```python
import requests
```

```python
from bs4 import BeautifulSoup
```

```python
url ='https://search.shopping.naver.com/best100v2/detail.nhn?catId=50000210'
```

```python
r = requests.get(url)
```

```python
r.text
```

```python
soup = BeautifulSoup(r.text, 'html.parser')
```

```python
soup
```

```python
print("++네이버 쇼핑 생활가전 순위++")
num=0
for i in goods:
    num +=1
    print(num, i['title'])
```
